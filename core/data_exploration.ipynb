```python
import chromadb
import h5py
import numpy as np
import json

# Config
CHROMA_DIR = "./chroma_db_mad_joint"
COLLECTION = "mad_joint_clip_L14"
JSON_PATH = r"C:\Users\nikhi\projects\AI-Video-Describer\MAD\annotations\MAD-v1\MAD_train.json"
H5_FRAMES = r"C:\Users\nikhi\projects\AI-Video-Describer\MAD\features\CLIP_L14_frames_features_5fps.h5"

# Load
client = chromadb.PersistentClient(path=CHROMA_DIR)
coll = client.get_collection(COLLECTION)
mad_data = json.load(open(JSON_PATH, "r"))

# Test frame
with h5py.File(H5_FRAMES, "r") as f:
    frame_emb = np.array(f["10142"][0]).astype(np.float32)
    frame_emb /= np.linalg.norm(frame_emb)

# Query with filter
result = coll.query(
    query_embeddings=[frame_emb.tolist()],
    n_results=15,
    where={"type": "text_caption"},
    include=["metadatas", "distances"]
)

print("Retrieved IDs:")
for i, (text_id, dist) in enumerate(zip(result["ids"][0], result["distances"][0])):
    clean_id = text_id.replace("text_", "")
    if clean_id in mad_data:
        print(f"{i+1}. {text_id} (dist={dist:.3f})")
        print(f"   {mad_data[clean_id]['sentence'][:100]}...")
    else:
        print(f"{i+1}. {text_id} ‚ùå NOT IN JSON")
```
